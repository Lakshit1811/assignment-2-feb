```python
numbers = []
for i in range(1, 101):
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        numbers.append(i)
        print(numbers)
```

```python
numbers = []
i = 1
while i <= 100:
    cube = i ** 3
    if cube % 4 == 0 or cube % 5 == 0:
        numbers.append(i)
    i += 1
print(numbers)
```
